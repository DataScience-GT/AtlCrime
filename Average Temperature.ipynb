```python
tmax = data['TMAX'].tolist()
tmin = data['TMIN'].tolist()

i = 0
while i < len(tmax):
    if str(tmax[i]) == 'nan' and str(tmin[i]) == 'nan':
        tmax[i] = 0
        tmin[i] = 0
    i += 1

data['TMAX'] = tmax
data['TMIN'] = tmin
```

```python
data.head(5)
```

```python
i = 0
while i < len(tmax):
    avg = (tmax[i] + tmin[i])/2
    data['TAVG'][i] = avg
    i += 1
```

```python
dates = data['IncidentToDate'].tolist()
for i in range(len(dates)):
    split = dates[i].split("/")
    if split[2] == '2013':
        data['IncidentToDate'][i] = '2013'
    elif split[2] == '2014':
        data['IncidentToDate'][i] = '2014'
    elif split[2] == '2015':
        data['IncidentToDate'][i] = '2015'
    elif split[2] == '2016':
        data['IncidentToDate'][i] = '2016'
    elif split[2] == '2017':
        data['IncidentToDate'][i] = '2017'
    elif split[2] == '2018':
        data['IncidentToDate'][i] = '2018'
```

```python
'''
IMPORTANT!! This is how you properly index and grab data that you want to look for. 
@Neha: You will want to change 'PRCP' to 'TAVG' or whatever you are looking for.
Note - this is only looking for data that occured in 2017. There is currently a slight mishap in the data which I need to fix, 
but for now pretend like there is no mistake in the data :)
'''
len(data[(data['TAVG'] < 100) & (data['TAVG'] > 30) & (data['IncidentToDate'] == '2017') & (data['OffenseCode'] != '9999W')])
```

```python
'''
This is gathering the number of occurrences of precipitation that was < 0.1, < 0.2, ...
@Neha: You will want to change this to reflect temperatures < 30, < 40, <50, ...
@Aditya: You will want to see if 'WT01' is set to 0 (for no fog) or 1 (for fog)
'''
crime_temp = []
for i in range(3, 11):
    num = i * 10
    crime_temp.append(len(data[(data['TAVG'] < num) & (data['IncidentToDate'] == '2017') & (data['OffenseCode'] != '9999W')]))
```

```python
'''
This is necessary to find the ranges of occurrences. For example, this will find me occurrences of crime when precipitation is
0.1 < x < 0.2, 0.2 < x < 0.3, ....
@All: You will need this.
'''
crime_temp_buckets = []
for i in range(len(crime_temp)):
    if i != 0:
        crime_temp_buckets.append(crime_temp[i] - crime_temp[i-1])
crime_temp_buckets
```

```python
'''
This plots a bar graph representing my data
@var data_y is the y-axis in a list. We calculated this above.
@var data_x is the x-axis values in a list.
'''
data_y = crime_temp_buckets
data_x = ['<40', '<50', '<60', '<70', '<80', '<90', '<100']

plt.bar(data_x, data_y, align='center', alpha=0.5)
plt.title("Crimes with Respect to Average Temperature in 2017")
plt.ylabel('Reported Cases of Crime')
plt.xlabel("Average temperature in Fahrenheit")
plt.show()
```

```python
'''
Normalizing data
each bucket = Total number of crimes/ Number of days with that temperature
@var data_y is the y-axis in a list. We calculated this above.
@var data_x is the x-axis values in a list.
'''
total_days_temp = []
for i in range(3, 11):
    num = i * 10
    total_days_temp.append(len(data[(data['TAVG'] < num) & (data['IncidentToDate'] == '2017')]))
```

```python
total_days_temp_buckets = []
for i in range(len(total_days_temp)):
    if i != 0:
        total_days_temp_buckets.append(total_days_temp[i] - total_days_temp[i-1])
total_days_temp_buckets
```

```python
normalized_data = []
for i in range(len(total_days_temp_buckets)):
    if (total_days_temp_buckets[i] != 0):
        normalized_data.append(total_days_temp_buckets[i]/crime_temp_buckets[i])
    else:
        normalized_data.append(0)
normalized_data
```

```python
'''
This plots a bar graph representing the normalized data
@var data_y is the y-axis in a list. We calculated this above.
@var data_x is the x-axis values in a list.
'''
data_y = normalized_data
data_x = ['<40', '<50', '<60', '<70', '<80', '<90', '<100']

plt.bar(data_x, data_y, align='center', alpha=0.5)
plt.title("Crimes with Respect to Average Temperature in 2017")
plt.ylabel('Crime')
#to exaggerate differences
#ylim(0.9, 1.0)
plt.xlabel("Average temperature in Fahrenheit")
plt.show()
```
